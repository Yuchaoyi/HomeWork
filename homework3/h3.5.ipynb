```python
import random
lst1 = random.sample(range(-200, 200), 100)
lst2 = random.sample(range(-2000, 2000), 1000)
lst3 = random.sample(range(-20000, 20000), 10000)
print(lst1)
print(lst2)
print(lst3)
```

```python
def charufun(lst):
    for i in range(1,len(lst)):
        t=lst[i]
        j=i-1
        while j>=0 and lst[j]>t:
            lst[j+1]=lst[j]
            lst[j]=t
            j=j-1
    return lst
```

```python
def xuanzefun(lst):
    for i in range(0,len(lst)-1):
        t=0
        minlst=lst[i]
        for j in range(i+1,len(lst)):
            if lst[j]<minlst:
                t=j
                minlst=lst[j]
        if minlst!=lst[i]:
            lst[t]=lst[i]
            lst[i]=minlst
    return lst
```

```python
def kuaisufun(lst):
    if len(lst)==1 or len(lst)==0:
        return lst
    left=[]
    right=[]
    for i in range(len(lst)-1):
        if lst[i]<lst[len(lst)-1]:
            left.append(lst[i])
        else:
            right.append(lst[i])
    return kuaisufun(left)+[lst[-1]]+kuaisufun(right)
```
