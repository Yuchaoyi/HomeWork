```python
import pandas as pd 
import numpy as np
```

```python
data=pd.read_csv('bike.csv')
```

```python
data.info()
```

```python
data.head(5)
```

```python
data.drop('id',axis=1,inplace=True)
```

```python
data.head(5)
```

```python
data=data[data.loc[:,'city']==1]
```

```python
data.drop('city',axis=1,inplace=True)
```

```python
data.head(5)
```

```python
data.reset_index(drop=True,inplace=True)
```

```python
data.head(5)
```

```python
data['hour'].info()
```

```python
data.loc[(data['hour']<6)|(data['hour']>18),'hour']=0
data.loc[(data['hour']>=6)&(data['hour']<=18),'hour']=1
```

```python
data['hour'].describe()
```

```python
ycol=data['y'].tolist()
```

```python
ycol=np.array(ycol)
```

```python
print(type(ycol))
```

```python
data.drop('y',axis=1,inplace=True)
```

```python
npdata=np.array(data)
```

```python
npdata.shape
```

```python
data.head(20)
```

```python
npdata1=npdata[:int(npdata.shape[0]*0.8),:]
```

```python
npdata2=npdata[int(npdata.shape[0]*0.8)::,:]
```

```python
ycol1=ycol[:int(npdata.shape[0]*0.8)]
ycol2=ycol[int(npdata.shape[0]*0.8):]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler_x=MinMaxScaler()
scaler_y=MinMaxScaler()
x_train_scaled=scaler_x.fit_transform(npdata1)
y_train_scaled=scaler_y.fit_transform(ycol1.reshape(-1,1))
x_test_scaled=scaler_x.transform(npdata2)
y_test_scaled=scaler_y.fit_transform(ycol2.reshape(-1,1))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
model=LinearRegression()
model.fit(x_train_scaled,y_train_scaled)
```

```python
from sklearn.preprocessing import StandardScaler
y_pred=model.predict(x_test_scaled)
# 对预测结果进行逆归一化
y_pred_original_scale = scaler_y.inverse_transform(y_pred)
print(y_pred_original_scale)
```

```python
mse=mean_squared_error(y_test_scaled,y_pred)
r2=r2_score(y_test_scaled,y_pred)
print(f"均方误差(MSE):{mse}")
print(f"R的平方（决定系数）：{r2}")
```

```python
# 打印模型的系数和截距
print(f"模型系数: {model.coef_}")
print(f"模型截距: {model.intercept_}")
```
